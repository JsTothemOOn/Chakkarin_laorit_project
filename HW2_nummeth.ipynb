```python
k = [0]*502; y = [0]*502; Nw = [0]*502; Nr = [0]*502; b = [1]*99 + [0.7]*403; h = [0]*502
d=1; s=0.1; k[1]=0.01; alpha = 0.5
Nw[1] = 100
Nr[1] = 100
for t in range(1,501):
     y[t] = (k[t]*Nw[t])**alpha
     k[t+1] = (1-d)*k[t]+s*y[t]
     Nr[t+1] = Nw[t]
     Nw[t+1] = b[t]*Nw[t]
     h[t] = y[t]/(Nw[t]+Nr[t])
```

```python
plt.figure(figsize=(20,10))
plt.title("h-happiness of economy")
plt.xlabel("timepath")
plt.ylabel("h")
plt.yticks(np.arange(0,0.1, 0.005))
plt.xticks(np.arange(0,200,10))
plt.plot(h[1:200])
plt.show 
```

```python
s_rates = np.linspace(0,1,100)
c500s = []
alpha = 0.9; d=0.1
y=[0]*502; k=[0]*502; 
k[1] = 0.1
for s in s_rates:
  for t in range(1,501):
    y[t] = k[t]**alpha
    k[t+1] = (1-d)*k[t] + s*y[t]
  c500s.append((1-s)*y[500])   

plt.plot(s_rates,c500s)
plt.xlabel("saving rates")
plt.ylabel("long-run consumption")
plt.title("long-run consumption vs Saving rate")
plt.show()
```

Saving rates that maximize long run consummption

```python
O = s_rates[np.argmax(c500s)]
print(f'Optimal saving rate = {O}')
```
