HW3 5) Solow model

```python
def optimal_srate(alpha):
  d=0.1
  y=[0]*502; k=[0]*502; 
  k[1] = 0.1
  c500s = []
  s_rate =  np.linspace(0,1,50)
  for s in s_rate :
    for t in range(1,501):
      y[t] = k[t]**alpha
      k[t+1] = (1-d)*k[t] + s*y[t]
    c500s.append((1-s)*y[500])
  optimal_saving_rate = s_rate[np.argmax(c500s)]
  return optimal_saving_rate
```

```python
#test function
optimal_srate(0.2)
```

```python
#a is level of alpha
a = np.linspace(0,1,50)
final_srate = []
for i in a:
    b = optimal_srate(i)
    final_srate.append(b)
plt.plot(a, final_srate)
plt.xlabel('Alpha level')
plt.ylabel('Optimal saving rate')
plt.show()
```

List comprehension version 

```python
c = list(map(optimal_srate, [i for i in np.linspace(0,1,50)]))
plt.plot(np.linspace(0,1,50), c)
plt.xlabel('Alpha level')
plt.ylabel('Optimal saving rate')
plt.show()
```

HW3 7) (Job assignment) 

```python
from itertools import permutations 
```

```python
output = [[43, 35, 31, 15, 44, 16],
       		 [39, 21, 39, 25, 33, 19],
       		 [38, 40, 25, 17, 49, 23],
       		 [20, 10, 12, 20, 38, 34],
       		 [45, 36, 11, 18, 35, 16],
       		 [13, 46, 48, 29, 14, 17]]
```

```python
#there are 720 possible solution 
index_num = []
job_func =permutations(list(range(0,6)))
for i in list(job_func):
    index_num.append(i)
```

```python
#change tuple in list to list in list
index_num1 = list(map(list,index_num))
index_num1
```

```python
#All possible total output 
y = []
for j in index_num1:
    x = 0
    for i in range(0,6):
        x += output[i][j[i]]
    y.append(x)
y
```

```python
#find index that maximize total output by using argmax and then find combination 
final_answer = index_num1[np.argmax(y)]
```

```python
#Finally, we got the answer
for index,i in enumerate(range(0,6)):
    print(f'People{str(index+1)} doing job{final_answer[i]+1} gain {output[i][final_answer[i]]}')
print(f'total output = {y[np.argmax(y)]}')
```
